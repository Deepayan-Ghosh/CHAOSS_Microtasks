# Contributor demographics proposed implementation
This document proposes a possible method to obtain contributor details that are essential for reporting of analysed projects based on Chaoss metrics, especially for the metric "Contributor demographics" presented here: https://github.com/chaoss/metrics/blob/master/1_Diversity-Inclusion.md;. 

### 1. The problem
The details required for reporting include age, gender, location, education etc. The problem is most of these details are not available from regular software development repositories like GitHub, git repos, etc.

### 2. Proposed solution
As a solution to the above problem, the required parameters have to be obtianed manually. I would suggest that the best way is to manually harvest these information through an input file whose path will be provided to `manuscript` as a command line argument. Thus, I propose the input file to be of `csv` format because:
1. It is easier to maintain by users, than other data exchange formats like json,yaml,etc
2. It can be directly read by python `pandas` and easily manipulated
3. The structure of required data doesn't conform to `key-value pair` or `nested` structure, so `csv` suits better

The columns will be `name`,`Date` for over-time comparison, `gender`, `education`, `age` and more, depending upon the information we need. For this demo implementation, I generate random sample data but in real life this will be obtained from repository contributors. Each column has some permitted values: e.g for `gender` they are `M for male, F for female, T for transgender`, for `education` they are `high school,bachelors,masters,Ph.D`, etc. These values will be determined based on the question given here: https://github.com/drnikki/open-demographics/tree/master/doc-source/questions

### 3. Comparing the data over time
The graphs at the end shows the distribution of contributors who contibuted over time, based on gender, age, and education. I have compared the data on a monthly basis for this demo, however for the real project, there would be options for: monthly, quarterly, semi-annually, anually. I have used line-chart for the comparative study. For example, while plotting the gender distribution, blue line represents male, red represents female, and green represents transgender. Having all the three lines in the same plot allows a better comparative study of the graph. The rise and fall of the lines allows us to analyse the number of males, females, transgenders for a particular month. Other charts like pie-chart cannot be used for over-time analysis, bar-charts does not make comparison between different similar factors easy. Hence I used line charts.

### N.B: `Location` needs some thought
Although `location` can be obtained from `user_location` field by querying enriched indices, the problem is that the user may not always provide his/her location in their GitHub profile, and then the value obtained from the index will be `null`. So, we have to decide whether to obtain the values through querying or through the `csv` file. `csv` file will point out to the users that they need to provide the location to us, and need not update their GitHub profile.

# Sample Code
First we import the required modules

````python
import matplotlib.pyplot as plt
import string
import random
import pandas as pd
import datetime
from collections import OrderedDict
````

An utility function for generating random dates

````python
def randomDate(start,end):
    return start + datetime.timedelta(
        # Get a random amount of seconds between `start` and `end`
        seconds=random.randint(0, int((end - start).total_seconds())),
    )
````

Here, I am generating 100 rows of random sample data, writing them to a `csv` file, to simulate the reading of `csv` file that would be needed.

````python
def create_sample_data():
    table = [
         # random name
        [''.join(random.choices(string.ascii_uppercase, k=5)),
         # random date
         randomDate(datetime.date(2015,1,1),datetime.date.today()),
         # allowed values for gender include M:male, F:female, T:transgender
         ''.join(random.choices(['M','F','T'])),
         # allowed values for education include: high school, bachelors, masters, Ph.D
         ''.join(random.choices(['High school','Bachelors','Masters','Ph.D'])),
         # random age
         random.randint(5,50)
        ] for _ in range(100)
    ]
    df = pd.DataFrame(table,columns = ['Name','Date of contribution','Gender','Education','Age'])
    df.to_csv('sample_data.csv',index=False)
````

This is the actual function which reads the `csv` data. 
It receives the name of file from the calling module, sorts them according to date (which makes plotting over time easier). To plot the data over time, we need `date` and `gender,age,education` data. So we make a dict with `date` as key, and the other info entirely as a dict, as the value for `date` key. I am comparing the data monthly, so the `date` values will be of the format `Month-Year`. So the function creates `buckets`, where each `bucket` (in `buckets`) is a month (key) and the value for each bucket is a dictionary having information about gender distribution of contributors for that month, that is, number of male,female and transgender contributors for that month. Similarly for education, number of high school contributors, number of bachelors, number of masters, etc. For age, we maintain age of oldest contributor, age of youngest contributor; `total_age` and `count` are used for calculating average age during plotting. Thus `buckets` is a nested dictionary.

```
buckets = {
    month: {
             'gender' : {'M':2,'F':0,'T':0 },
             'education': { 'High school':1,'Bachelors':4,'Masters':0,'Ph.D':0 },
             'age' : {'max':34,'min':17,'total_age':51, 'count':2 }
         }
    ............................

```

````python
# This is an individual activity metric which belongs to diverity inclusion metric
# and, hence will be called from a function named diversity_inclusion_metric()
def contributor_demographics(filename):
    df = pd.read_csv(filename)
    df['Date of contribution'] = pd.to_datetime(df['Date of contribution'])
    df = df.sort_values(['Date of contribution'],ascending=True)
    print("The table:\n")
    print(df)
    
    # OrderedDict helps to preserve order, and does not sort the contents according to `key` that is month name
    buckets = OrderedDict()
    for _, row in df.iterrows():
        bucket = row['Date of contribution'].strftime("%b") + "-" + row['Date of contribution'].strftime("%Y")
        gender = row['Gender']
        education = row['Education']
        age = row['Age']
        if bucket not in buckets:
            buckets[bucket] = { 
                                'gender' : {'M':0,'F':0,'T':0 },
                                'education': { 'High school':0,'Bachelors':0,'Masters':0,'Ph.D':0 },
                                'age' : {'max':0,'min':9999,'total_age':0, 'count':0 },
                            }
        buckets[bucket]['gender'][gender] += 1
        buckets[bucket]['education'][education] += 1
        
        if age > buckets[bucket]['age']['max']:
            buckets[bucket]['age']['max'] = age
        if age < buckets[bucket]['age']['min']:
            buckets[bucket]['age']['min'] = age
        buckets[bucket]['age']['total_age'] += age
        buckets[bucket]['age']['count'] += 1

    # plot the data
    plot(buckets)
````

This function takes the data and plots gender-data over time, education distribution over time, and age distribution over time as three subplots of appropriate sizes with proper legends.

````python
def plot(buckets):
    gender_info = []
    # months will be the values on X-axis
    x = buckets.keys()
    males,females,third = [],[],[]
    skl, bach, mast, phd = [],[],[],[]
    max_age,min_age,avg_age = [],[],[]
    for each in x:
        # append the number of males, females, transgenders for each month 
        # so that they can be used in Y-axis while plotting
        males += [buckets[each]['gender']['M']]
        females += [buckets[each]['gender']['F']]
        third += [buckets[each]['gender']['T']]
        
        # Similarly append the number of high school, bachelor, masters, and ph.d contributors for each month 
        skl += [buckets[each]['education']['High school']]
        bach += [buckets[each]['education']['Bachelors']]
        mast += [buckets[each]['education']['Masters']]
        phd += [buckets[each]['education']['Ph.D']]
        
        # append the max, min, and calculated average age for each month
        max_age += [buckets[each]['age']['max']]
        min_age += [buckets[each]['age']['min']]
        avg_age += [buckets[each]['age']['total_age'] / buckets[each]['age']['count']]
        
    fig = plt.figure(figsize = (25,30))
    fig.suptitle("Contributor Demographics",fontsize=28)
    
    # Plot number 1
    ax1 = fig.add_subplot(311)
    ax1.plot(x, males, label='Male', color='b', marker='o')
    ax1.plot(x, females, label='Female', color='r', marker='o')
    ax1.plot(x, third, label='Transgender', color='g', marker='o')
    plt.xticks(rotation=70)
    plt.title("Gender over time", fontsize=20)
    handles, labels = ax1.get_legend_handles_labels()
    lgd = ax1.legend(handles, labels, loc='upper center', bbox_to_anchor=(1,1))
    ax1.grid(True)
    
    # Plot number 2
    ax2 = fig.add_subplot(312)
    ax2.plot(x, skl, label='High school', color='b', marker='o')
    ax2.plot(x, bach, label='Bachelors', color='r', marker='o')
    ax2.plot(x, mast, label='Masters', color='g', marker='o')
    ax2.plot(x, phd, label='Ph.D', color='c', marker='o')
    plt.xticks(rotation=70)
    plt.title("Education over time", fontsize=20)
    handles, labels = ax2.get_legend_handles_labels()
    lgd = ax2.legend(handles, labels, loc='upper center', bbox_to_anchor=(1,1))
    ax2.grid(True)
    
    # Plot number 3
    ax3 = fig.add_subplot(313)
    ax3.plot(x, max_age, label='Maximum age', color='b', marker='o')
    ax3.plot(x, min_age, label='Minimum age', color='r', marker='o')
    ax3.plot(x, avg_age, label='Average age', color='g', marker='o')
    plt.xticks(rotation=70)
    plt.title("Age comparison over time", fontsize=20)
    handles, labels = ax3.get_legend_handles_labels()
    lgd = ax3.legend(handles, labels, loc='upper center', bbox_to_anchor=(1,1))
    ax3.grid(True)
````
